## Import Library


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style of the plots
sns.set_style("whitegrid")
```

```python
df = pd.read_csv('product_sales_cleaned.csv')
df.head()
```

## Data Analysis
### Single Variable Analysis

```python
def plot_single_subplots(df):
    
    # Setting up for multiple subplots and adjusting the layout
    fig, axes = plt.subplots(3, 2, figsize=(16, 12))
    plt.tight_layout(pad=5.0)

    # the distribution of 'week'
    sns.countplot(ax=axes[0, 0], x='week', data=df, palette='viridis')
    axes[0, 0].set_title('Distribution of Sales by Week')
    axes[0, 0].set_xlabel('Week')
    axes[0, 0].set_ylabel('Count')

    # the distribution of 'nb_sold'
    sns.histplot(ax=axes[0, 1], x='nb_sold', data=df, bins=30, kde=True, color='skyblue')
    axes[0, 1].set_title('Distribution of Number of Products Sold')
    axes[0, 1].set_xlabel('Number of Products Sold')
    axes[0, 1].set_ylabel('Frequency')

    # Plotting the distribution of 'revenue'
    sns.histplot(ax=axes[1, 0], x='revenue', data=df, bins=30, kde=True, color='green')
    axes[1, 0].set_title('Distribution of Revenue')
    axes[1, 0].set_xlabel('Revenue')
    axes[1, 0].set_ylabel('Frequency')

    # Plotting the distribution of 'years_as_customer'
    sns.histplot(ax=axes[1, 1], x='years_as_customer', data=df, bins=30, kde=True, color='orange')
    axes[1, 1].set_title('Distribution of Years as Customer')
    axes[1, 1].set_xlabel('Years as Customer')
    axes[1, 1].set_ylabel('Frequency')

    # Plotting the distribution of 'nb_site_visits'
    sns.histplot(ax=axes[2, 0], x='nb_site_visits', data=df, bins=30, kde=True, color='purple')
    axes[2, 0].set_title('Distribution of Number of Site Visits')
    axes[2, 0].set_xlabel('Number of Site Visits')
    axes[2, 0].set_ylabel('Frequency')

    # Count of Sales Methods Used
    sns.countplot(ax=axes[2, 1], data=df, x='combined_sales_method', palette='coolwarm')
    axes[2, 1].set_title('Count of Sales Methods Used')
    axes[2, 1].set_xlabel('Sales Method')
    axes[2, 1].set_ylabel('Count')

    # Adding count numbers on top of the bars for the Sales Method graph
    for p in axes[2, 1].patches:
        axes[2, 1].annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), 
                            ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                            textcoords='offset points')


    plt.show()
```

### Multiple Variables Analysis

```python
# Relationship between Sales Method and Revenue
def plt_mul_var_1(df):
    plt.figure(figsize=(10, 6))
    ax = sns.barplot(x='combined_sales_method', y='revenue', data=df, ci=None, palette='pastel')

    # Adding the average revenue numbers on top of the bars
    for p in ax.patches:
        ax.annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2., p.get_height()), 
                    ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                    textcoords='offset points')

    plt.title('Revenue by Sales Method')
    plt.xlabel('Sales Method')
    plt.ylabel('Average Revenue')
    plt.show()
```

```python
# Revenue by Sales Method over Time
def plt_mul_var_2(df):
    plt.figure(figsize=(14, 8))
    sns.lineplot(data=df, x='week', y='revenue', hue='combined_sales_method', marker='o')
    plt.title('Revenue by Sales Method Over Time')
    plt.xlabel('Week')
    plt.ylabel('Average Revenue')
    plt.legend(title='Sales Method')
    plt.show()
```
